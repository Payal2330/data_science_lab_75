```python
def infer_decision_short(facts, knowledge_base):
    for rule in knowledge_base:
        conditions_met = True
        for key, value in rule['conditions'].items():
            if key not in facts or facts[key].lower() != value.lower():
                conditions_met = False
                break
        if conditions_met:
            return rule['decision']
    return 'No decision could be made.'

knowledge_base_short = [
    {'conditions': {'weather': 'rainy'}, 'decision': 'Bring umbrella'},
    {'conditions': {'weather': 'sunny'}, 'decision': 'No umbrella'},
    {'conditions': {'weather': 'cloudy', 'temperature': 'cold'}, 'decision': 'Wear jacket'}
]

# Example usage:
facts1 = {'weather': 'Rainy'}
decision1 = infer_decision_short(facts1, knowledge_base_short)
print(f"Facts: {facts1} -> Decision: {decision1}")

facts2 = {'weather': 'Sunny', 'temperature': 'hot'}
decision2 = infer_decision_short(facts2, knowledge_base_short)
print(f"Facts: {facts2} -> Decision: {decision2}")

facts3 = {'weather': 'cloudy', 'temperature': 'cold'}
decision3 = infer_decision_short(facts3, knowledge_base_short)
print(f"Facts: {facts3} -> Decision: {decision3}")

facts4 = {'weather': 'cloudy', 'temperature': 'warm'}
decision4 = infer_decision_short(facts4, knowledge_base_short)
print(f"Facts: {facts4} -> Decision: {decision4}")
```
