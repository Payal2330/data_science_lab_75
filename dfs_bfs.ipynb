```python
def dfs(graph, start_node):
    visited = set()
    stack = [start_node]

    while stack:
        node = stack.pop() # Get the last element (LIFO)
        if node not in visited:
            print(node, end=' ')
            visited.add(node)
            # Add unvisited neighbors to the stack
            # Reverse order to process smaller neighbors first if needed, otherwise order doesn't strictly matter for correctness
            for neighbor in reversed(graph.get(node, [])):
                if neighbor not in visited:
                    stack.append(neighbor)

# Example Graph (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("DFS Traversal (starting from A):")
dfs(graph, 'A')
```

```python
from collections import deque

def bfs(graph, start_node):
    visited = set()
    queue = deque([start_node]) # Use a deque for efficient pop from left
    visited.add(start_node)

    while queue:
        node = queue.popleft() # Get the first element (FIFO)
        print(node, end=' ')

        # Add unvisited neighbors to the queue
        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

# Example Graph (Adjacency List) - using the same graph as DFS for comparison
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("\n\nBFS Traversal (starting from A):")
bfs(graph, 'A')
```
